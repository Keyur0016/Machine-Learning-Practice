```python
import pandas as pd 
import numpy as np 
```

```python
# Load Dataset

Dataset = pd.read_csv('./Dataset/NaiveByes.csv')
Dataset.head(5)
```

```python
# Convert this String Data column to Numericvalue Data 

from sklearn.preprocessing import LabelEncoder

Le_Outlook = LabelEncoder()
Le_Temperature = LabelEncoder() 
Le_Humidity = LabelEncoder()
Le_Wind = LabelEncoder() 
```

```python
# Convert Outlook column to numeric value coluumn 

# sunny = 2, overcast = 0, rain = 1

Dataset['Le_Outlook'] = Le_Outlook.fit_transform(Dataset['Outlook'])
Dataset
```

```python
# Convert Temperature and other columns to numeric value columns 

Dataset['Le_Temperature'] = Le_Temperature.fit_transform(Dataset['Temperature'])
Dataset['Le_Humidity'] = Le_Humidity.fit_transform(Dataset['Humiidity'])
Dataset['Le_Wind'] = Le_Wind.fit_transform(Dataset['Wind'])

Dataset
```

```python
# Value 

# For Outlook column => sunny=2, overcast=0, rain=1 
# For Temperature column => hot=1, miild=2, cool=0 
# For Humidiity column => high =0, normal=1
# For Wind column => weak=1, strong=0 

Target_dataFrame = Dataset['PlayTenniis']
Target_dataFrame
```

```python
Dataset = Dataset.drop(['Outlook','Temperature','Humiidity','Wind','PlayTenniis'], axis=1)
Dataset
```

```python
# Train,test and Split Data 

from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(Dataset,Target_dataFrame,test_size=0.2)
```

```python
# Now Load NaiveByes Model 

from sklearn.naive_bayes import GaussianNB

Model = GaussianNB() 
Model.fit(X_train,Y_train)
```

```python
# Now Make Prediction

Prediction = Model.predict(np.array([[2,0,0,0]]))
Prediction
```
